Código desenvolvido durante um intensivão de python.

Destinado a automatização de tarefas. Sendo através do mesmo realizada a obtenção da cotação do ouro, dolar e euro, e a alteração de dados em uma planilha no excel baseados no valor das moedas no momento.

Para o funcionamento do código é necessária a instalação do selenium e do WebDriver.

```python
# !pip install selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

# Criação do navegador
navegador = webdriver.Chrome()
```

```python
navegador.get("https://www.google.com.br/")

# Obter cotação do Dolar
navegador.find_element('xpath', 
                       '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("Cotação dolar")

#Usar teclas
navegador.find_element('xpath', 
                       '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

#Obter atributo da página WEB
cotacaoDolar = navegador.find_element('xpath', 
                       '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
```

```python
#Obter cotação do euro 

navegador.get("https://www.google.com.br/")

navegador.find_element('xpath', 
                       '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("Cotação Euro")

#Usar teclas
navegador.find_element('xpath', 
                       '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

#Obter atributo da página WEB
cotacaoEuro = navegador.find_element('xpath', 
                       '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
```

```python
# Obter a cotação do ouro

navegador.get("https://www.melhorcambio.com/ouro-hoje#:~:text=O%20valor%20do%20grama%20do,em%20R%24%20316%2C39.")

cotacaoOuro = navegador.find_element('xpath', '//*[@id="comercial"]').get_attribute('value')

cotacaoOuro = cotacaoOuro.replace(',', '.')

#fechando o navegador

navegador.quit()
```

```python
#Visualização dos valores extraídos

print(cotacaoDolar)
print(cotacaoEuro)
print(cotacaoOuro)
```

```python
#Tratar as informações

import pandas as pd
import openpyxl
import numpy

tabela = pd.read_excel("tabela_produtos.xlsx")
display(tabela)
```

```python
tabela.loc[tabela["Moeda"] == "Dolar", "Cotação "] = float(cotacaoDolar)
tabela.loc[tabela["Moeda"] == "Ouro", "Cotação "] = float(cotacaoOuro)
tabela.loc[tabela["Moeda"] == "Euro", "Cotação "] = float(cotacaoEuro)

display(tabela)
```

```python
#Recalcular os preços

tabela["Preço de Compra"] = tabela["Cotação"] * tabela["Preço Original"]


display(tabela)
```

```python
tabela["Preço de Venda"] = tabela["Preço de Compra"] * tabela["Margem "]


display(tabela)
```

```python
#Formatação dos valores + exportação da base para o excel

tabela.to_excel("Produtos_Novo.xlsx", index = False)
```
